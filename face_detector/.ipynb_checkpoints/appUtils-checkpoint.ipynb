```python
import cv2 as cv
import threading
import time
```

```python
ESP32_STREAM_URL = "http://192.168.100.251/stream"

cap_lock = threading.Lock()
cap = None
```

```python
def release_video_capture():
    """
    Giải phóng capture hiện tại (nếu có).
    """
    global cap
    with cap_lock:
        if cap is not None and cap.isOpened():
            cap.release()
            cap = None
            print("[INFO] Released VideoCapture")
```
